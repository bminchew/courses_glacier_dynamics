## Exercise 1: Effective Viscosity of Ice

Glen's flow law relates the deviatoric stress to the strain rate via a power-law rheology. The effective viscosity is defined as

$$\eta = \frac{1}{2} A^{-1/n} \dot{\varepsilon}_e^{\,(1-n)/n}$$

where $\dot{\varepsilon}_e$ is the effective strain rate. For $n > 1$, the viscosity *decreases* with increasing strain rate — ice is a shear-thinning fluid.

## Exercise 4: Profile Curvature Diagnostic

The shape of an ice sheet's surface profile reveals information about the dominant stress regime. The SIA predicts a convex profile everywhere (as in the Vialov solution), while membrane/lateral stresses tend to produce concave profiles near the margins.

The sign of $d^2(h/h_0)/d(x/L)^2$ tells us:
- **Concave** ($d^2h/dx^2 > 0$): profile curves upward — consistent with membrane stresses or sliding.
- **Convex** ($d^2h/dx^2 < 0$): profile curves downward — consistent with SIA-like deformational flow.
